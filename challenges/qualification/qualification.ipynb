## Loading and visualization of the traces

The following code loads the previously captured power traces. These traces were obtained using a ChipWhisperer Nano and represent the device's power consumption while different single character passwords were attempted as inputs.

The objective of this analysis is to examine these variations in power consumption to identify patterns that allow us to deduce the correct first character of the password. This is known as a Side-Channel Attack, specifically, a Power Analysis Attack.

These traces can be used as reference in your qualification report, but a full solution does not need to be implemented.

```python
import numpy as np
import matplotlib.pyplot as plt

try:
    input_file = 'jupyter/user/qualification/side_channel_traces_char1.npz'
    data = np.load(input_file)
except FileNotFoundError:
    print(f"Error: File not found at '{input_file}'.")
    print("Make sure you run the capture and save script first.")
    exit()

loaded_traces = data['traces']
loaded_plaintexts = data['plaintexts']

print(f"âœ… Data loaded successfully from '{input_file}'.")
print(f"Found {len(loaded_traces)} traces to visualize.")

for i in len(loaded_traces)
    trace = loaded_traces[i]
    plaintext = loaded_plaintexts[i]

    tested_char = chr(plaintext[0])

    plt.figure(figsize=(12, 3))
    plt.title(f"Trace for tested character: '{tested_char}'")
    
    plt.plot(trace[:500])
    
    plt.xlabel("Time Samples")
    plt.ylabel("Power Consumption (Voltage)")
    plt.grid(True)
    plt.show()
```
